```python
import pandas as pd
#感情付与データ
data = pd.read_csv("gakusyuu_honban.csv")
#感情付与なしデータ
# data = pd.read_csv("../../kanjyou/multi_pos/kanjyou_huyo_multi_weight2.csv")
data
```

```python
data = data[["発話者2","応答者2","対話行為","感情"]]
data.rename(columns={'発話者2': '発話者', '応答者2': '応答者'}, inplace=True)
data.head()
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("rinna/japanese-gpt2-small")
```

```python
import pandas as pd

with open('arinasi1.txt', 'w') as datadata:
    
    #tmp = []
    for x, y in zip(data["発話者"], data["応答者"]):
            inp_tokens = tokenizer.tokenize(x)[:100]
            inp = "".join(inp_tokens).replace('▁', '')
            out_tokens = tokenizer.tokenize(y)[:100]
            out = "".join(out_tokens).replace('▁', '')

            data = "<s>" + inp + "[SEP]" + out + "</s>"
            #tmp.append(data)
            txt = "".join(data)
            datadata.write(txt + '\n')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

with open('arinasi1.txt', 'r') as f:
    data = f.readlines()
    
x = data
# 訓練データとバリデーションデータの分割
x_train, x_val = train_test_split(x, test_size=0.2, random_state=42)

y = x_val 

y_val, y_test= train_test_split(y, test_size=0.5, random_state=42)


print(len(x_train))
print(len(y_val))
print(len(y_test))

with open('arinasi1_train.txt', 'w', encoding='utf-8') as train:
    train.writelines(x_train)
with open('arinasi1_val.txt', 'w', encoding='utf-8') as val:
    val.writelines(y_val)
with open('arinasi1_test.txt', 'w', encoding='utf-8') as test:
    test.writelines(y_test)
```
