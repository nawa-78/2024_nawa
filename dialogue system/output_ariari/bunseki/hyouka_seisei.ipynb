```python
import pandas as pd
df = pd.read_excel("kekka_bunseki.xlsx")
```

```python
df_kazu = df["応答数"]
df 
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# サンプルのDataFrameを作成する


sns.boxplot(data = df_kazu)

# グリッドを表示する
plt.grid(True)

# グラフを表示する
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

print(len(df[df['応答数']== 30]))

plt.hist(df['応答数'], bins=6,range=(0, 30), color='skyblue', edgecolor='black')

# グリッドを表示する
plt.grid(True)

# x軸とy軸にラベルを付ける
plt.xlabel('Repetition Count')
plt.ylabel('Frequency')

# タイトルを付ける


plt.savefig('histogram_example.png', dpi=300)

# ヒストグラムを表示する
plt.show()
```
