```python
#発話者→応答者のペア
all_files = glob.glob( "*.xlsx")
dfv = pd.DataFrame()
for filename in all_files:

    target = '.'
    idx = filename.find(target)
    r = filename[:idx]

    df__1 = pd.DataFrame()

    file_name = pd.read_excel(filename)
    

    kaiwa = file_name[[ "発話ID","話者", "発話文", "話者交代", "対話行為"]]
    wasya1 = []
    wasya2 = []
    hatu1 = []
    hatu2 = []
    dai1 =[]
    daidai1 = 0
    dai2 =[]
    daidai2 = 0
    ki = []
    gu = []
    ma =[]
    fu =[]
    koui1 = []
    koui2 = []

    wa = 0


    for p,i, j, k, l in zip(kaiwa["発話ID"], kaiwa['話者'], kaiwa['発話文'], kaiwa['話者交代'], kaiwa["対話行為"]):


        if k == "1":
            wa = wa + 1

            if wa % 2 == 1:#話者の判別
                result1 = "".join(hatu1)
                wasya1.append(result1)
                dai1.append(daidai1)
                ki.append(p)
                fu.append(i)
                hatu1 = []
                hatu1.append(j)
                koui1.append(l)
                daidai1 = l




            elif wa % 2 == 0:
                result2 = "".join(hatu2)
                wasya2.append(result2)
                dai2.append(daidai2)
                gu.append(p)
                ma.append(i)
                hatu2 = []
                hatu2.append(j)
                koui2.append(l)
                daidai2 =l


        elif k == "0":
            if wa % 2 == 1:

                hatu1.append(j)
                daidai1 = l


            elif wa % 2 == 0:
                hatu2.append(j)
                daidai2 = l

    result1 = "".join(hatu1)
    wasya1.append(result1)
    dai1.append(daidai1)

    result2 = "".join(hatu2)
    wasya2.append(result2)
    dai2.append(daidai2)
    
    
    del wasya1[0]
    #del wasya1[1]
    del wasya2[0]
    del dai1[0]
    del dai2[0]

    
    df1 = pd.DataFrame(wasya1,columns=['発話者'])
    bun_1 = df1['発話者']
    bun_1 = bun_1.str.replace('（.*?）', '')
    bun_1 = bun_1.str.replace('\(.*?\)', '')
    bun_1 = bun_1.str.replace('<笑い>', '')
    bun_1 = bun_1.str.replace('＜笑い＞', '')
    bun_1 = bun_1.str.replace('【.*?】', '')
    bun_1 = bun_1.str.replace('＊', '')    

    df1["対話行為"]= pd.DataFrame(dai1)
    df1["対話行為2"]= pd.DataFrame(koui1)
    df1["ID"] =pd.DataFrame(ki)
    df1["話者"] =pd.DataFrame(fu)
    

    df2 = pd.DataFrame(wasya2,columns=['応答者'])
    bun_2 = df2['応答者']
    bun_2 = bun_2.str.replace('（.*?）', '')
    bun_2 = bun_2.str.replace('\(.*?\)', '')
    bun_2 = bun_2.str.replace('<笑い>', '')
    bun_2 = bun_2.str.replace('＜笑い＞', '')
    bun_2 = bun_2.str.replace('【.*?】', '')
    bun_2 = bun_2.str.replace('＊', '')
    
    df2 = df2.drop(["応答者"], axis = 1)
    df2 = pd.concat([bun_2,df2], axis = 1)
    
    df2["対話行為"]= pd.DataFrame(dai2)
    df2["対話行為2"]= pd.DataFrame(koui2)
    df2["ID"] =pd.DataFrame(gu)
    df2["話者"] =pd.DataFrame(ma)


    df__1 = pd.concat([bun_1,df2], axis = 1)
    df__1 = df__1.dropna() 
    df__1

    dfv = pd.concat([dfv, df__1], axis=0)

#dfv 
dfv.to_excel( "hatuwasya_saki_all.xlsx",index = False, encoding = "UTF-8")
#dfv.to_csv( "ketugou_2_all.csv",index = False, encoding = "UTF-8")
#print(len(dfv["対話行為"]))
```

```python
import re
with open('bun_all2.csv', 'r', encoding='utf-8') as f:
    text = f.read()
    # かっこを全て削除
    text = re.sub(r'（.*?）', '', text)
    text = re.sub(r'\(.*?\)', '', text)
    # <笑い> を全て削除
    text = re.sub(r'<笑い>', '', text)
    text = re.sub(r'＜笑い＞', '', text)
    text = re.sub(r'【.*?】', '', text)
    text = re.sub(r'＊', '', text)

with open('bun2.csv', 'w', encoding='utf-8') as f:
    f.write(text)
#文の修正    
```
